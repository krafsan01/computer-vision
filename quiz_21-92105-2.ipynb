```python
def bubbleSort(array):  
    for i in range(len(array)):
        print('i=',i,end=' ')
        for j in range(0, len(array) - i - 1):
            print('\t j=',j)
            if array[j] > array[j + 1]:
                temp = array[j]
                array[j] = array[j+1]
                array[j+1] = temp
                print()
                print(f'Pushing >>>> {temp}')
                print()
            
            print(f'Data : {array}')
            
            


data = [11,2,12,43,35,76,324,1,-13]

bubbleSort(data)

print('Sorted Array in Ascending Order:')
print(data)
```

```python
data = [11,2,12,43,35,76,324,1,7]
```

```python
def quick_sort(array):
    print(f'Array : {array}')
    length = len(array)
    if length <=1:
        print('No sorting needed breaking logic')
        return array
    else:
        pivot = array.pop()
        print(f'Pivot is now {pivot}')
        
    items_greater = []
    items_lower = []
    
    for item in array:
        
        if item > pivot :
            items_greater.append(item)
            print(f'{item}  is greater than {pivot}')
            print(f'List of greater than {pivot} itmes : {items_greater}')
        else:
            items_lower.append(item)
            print(f'{item}  is smaller than {pivot}')
            print(f'List of lower than {pivot} itmes : {items_lower}')
    print('----------------------------------')
    print(f'List of greater than {pivot} itmes : {items_greater}')
    print(f'List of lower than {pivot} itmes : {items_lower}')
    print('----------------------------------')
    
    print(quick_sort(items_lower) + [pivot] + quick_sort(items_greater))
    return quick_sort(items_lower) + [pivot] + quick_sort(items_greater)
             
        
        
    
result = quick_sort(data)
print(result)
```
